```python
#conclusions 
#1) new sectors or 70 onwards have more flats while old sectors have more houses
```

```python
plt.figure(figsize=(12, 30))

# Creating the pivot table and sorting by 'price' (mean)
pivot_table = pd.pivot_table(df, index='sector', values='price', aggfunc='mean').sort_values(by='sector', ascending=False)

# Plot the sorted heatmap
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')

plt.title("Average Price by Sector")
plt.show()
```

```python
plt.figure(figsize=(12, 30))

# Creating the pivot table and sorting by 'price' (mean)
pivot_table = pd.pivot_table(df, index='sector', values='price_per_sqft', aggfunc='mean').sort_values(by='sector', ascending=False)

# Plot the sorted heatmap
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')

plt.title("Average Price by Sector")
plt.show()
```

```python
plt.figure(figsize=(12, 30))

# Creating the pivot table and sorting by 'price' (mean)
pivot_table = pd.pivot_table(df, index='sector', values='luxury_score', aggfunc='mean').sort_values(by='sector', ascending=False)

# Plot the sorted heatmap
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')

plt.title("Average Price by Sector")
plt.show()
```
