# removing unimportant columns

```python
df.drop(columns=['link','property_id'],inplace=True)
```

```python
df.rename(columns={'area':'price_per_sqft'},inplace = True)
```

# society column

```python
df['society'].value_counts()
```

```python
df['society'].nunique()
```

```python
import re
df['society'] = df['society'].str.replace(r'\d+\.\d+\s*★', '', regex=True).str.strip()
```

```python
df['society'].nunique()
```

```python
df['society'] = df['society'].apply(lambda x : x.lower() if isinstance(x, str) else x)
```

```python
df['society'].nunique()
```

```python
df = df[~df['society'].isnull()]
df.shape
```

# price column

```python
df['price'].value_counts().shape
```

```python
# removing values which are 'Price on Request'
df = df[~((df['price']=='Price on Request') | (df['price'].isnull()))]
```

```python
def handling_price(price):
    if isinstance(price, str):
        parts = price.split()  
        value = float(parts[0])  
    
        if parts[1] == 'Lac':  
            return round(value / 100, 2)
        else:
            return round(value, 2)
            
    else:
        return price

df['price'] = df['price'].apply(handling_price)
    
```

```python
df.sample(5)
```

# price per sqft

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split().str[1].str.split('/').str[0].str.replace(',', '').str.strip().astype(float)
```

# bed room, bath room and balcony

```python
df['bedRoom'].value_counts()
```

```python
df['bedRoom'] = df['bedRoom'].str.split().str[0].astype(int)
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'] = df['bathroom'].str.split().str[0].astype(int)
```

```python
df['balcony'].value_counts()
```

```python
df['balcony'] = df['balcony'].str.split().str[0].str.replace('No','0')
```

# additional rooms column

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'] = df['additionalRoom'].fillna('not available')
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

# floor numbers

```python
df['floorNum'].value_counts()
```

```python
df['floorNum'].isnull().sum()
```

```python
df['floorNum'] = df['floorNum'].str.split().str.get(0).str.replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract('(\d+)')
```
