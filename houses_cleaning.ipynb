# basic preprocessing

```python
df.drop(columns=['link','property_id'], inplace=True)
```

# price

```python
df['price'].isnull().sum()
```

```python
df['price'].value_counts()
```

```python
df = df[~((df['price']=='Price on Request') | (df['price'].isnull()))]
```

```python
df.shape
```

```python
def handling_price(price):
    parts = price.split() 
    value = float(parts[0])
    type = parts[1]
    if type == 'Lac':
        return round(value/100,2)

    else:
        return round(value,2)
```

```python
df['price'] = df['price'].apply(handling_price)
```

# Society

```python
df['society'].value_counts()
```

```python
df['society'] = df['society'].str.replace(r'\d+\.\d+\s*★', '', regex=True).str.strip()
df['society'] = df['society'].apply(lambda x : x.lower() if isinstance(x, str) else x)
```

# rate column

```python
df.rename(columns={'rate':'price_per_sqft'},inplace = True)
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split(' ').str.get(1).str.split('/').str.get(0).str.strip().str.replace(',','').astype(float)
```

```python
df['area'] = df['price']*10000000/df['price_per_sqft']
```

```python
df.sample(5)
```

# bedroom, bathroom and balcony column

```python
df['bedRoom'].value_counts()
```

```python
df['bedRoom'].isnull().sum()
```

```python
df['bedRoom'] = df['bedRoom'].str.split().str.get(0).astype(int)
```

```python
df['bathroom'].isnull().sum()
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'] = df['bathroom'].str.split().str.get(0).astype(int)
```

```python
df['balcony'].isnull().sum()
```

```python
df['balcony'].value_counts()
```

```python
df['balcony'] = df['balcony'].str.split().str.get(0).str.replace('No','0')
```

# additional room

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
df['additionalRoom'] = df['additionalRoom'].fillna('not available')
```

# floor numbers

```python
df.rename(columns = {"noOfFloor":"floorNum"},inplace = True)
```

```python
df['floorNum'].isnull().sum()
```

```python
df['floorNum'].value_counts()
```

```python
df['floorNum'] = df['floorNum'].str.split().str.get(0)
```

```python
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('NA', inplace=True)
```
